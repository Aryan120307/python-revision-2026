```python
# property - additive property
#(a+b).T=a.T+b.T
#multiplicative property
#(a*b).T=(a.T)*(b.t)
```

```python
# Create sample matrices
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Calculate sides
right_side = (a.T) + (b.T)
left_side = (a + b).T

# Check equality
print("Is (a+b).T == (a.T) + (b.T) equal : \n", np.array_equal(left_side, right_side))
# Output: True
```

```python
# Create sample matrices
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Calculate sides for the correct identity (AB).T == B.T @ A.T
left_side = (a @ b).T
right_side = (b.T) @ (a.T) # Note the reversed order: b.T @ a.T

# Check equality
print("Is (a @ b).T == (b.T) @ (a.T) equal:", np.array_equal(left_side, right_side))
```
