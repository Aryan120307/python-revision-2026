```python
#outliers
sns.boxplot(df['Population'])
plt.show()
```

```python
q1=df['Population'].quantile(0.25)
q3=df['Population'].quantile(0.75)
```

```python
iqr= q3-q1
```

```python
lb=q1-1.5*iqr
ub=q3+1.5*iqr
```

```python
df=df[(df['Population']>=lb) & (df['Population']<=ub)]
df
```

```python
#variance
df['Population'].var()
```

```python
#standard deviation
df['Population'].std()
```

```python
#correlation - it will only work for numeric data
df.corr(numeric_only=True)
```
