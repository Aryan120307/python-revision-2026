```python
# ENCODING - it is of three types
# 1. label encoding = will use alphabetical order
#     apple=0
#     cherry=3
#     banana=2
# 2. ordinal encoding = will use order provided by user 
#     low=0
#     high=2
#     med=1
# 3. one hot encoding = it will create extra columns and then it will encode it 
#     status           single    married   relationship
#     single              1        0           0
#     married             0        1           0
#     relationship        0        0           1
```

```python
# ENCODING - CONVERTING CATEGORICAL DATA TO NUMERICAL DATA 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
    if df[i].dtype=='object':
        df[i]=le.fit_transform(df[i])
```

```python
# STEP:5 (DATA SPLITTING FOR MODEL BUILDING)
x=df.drop("Churn",axis=1) # independent variable
y=df['Churn'] # dependent variable
```

```python
# divide my data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) # random state = to shuffle the data , test size= 0.2 mean 20% data goes for testing 
```

```python
# STEP:6 MODEL TRAINING AND TESTING
from sklearn.ensemble import RandomForestClassifier
model_1=RandomForestClassifier()

# TRAIN THE MODEL 
model_1.fit(x_train,y_train)

# ASK THE MODEL TO PREDICT 
y_pred= model_1.predict(x_test)

# EVALUATING THE PERFORMANCE BY COMPARING THE PREDICTED VALUE WITH THE ACTUAL VALUE
from sklearn.metrics import *
print("Accuracy Score of the model is :", accuracy_score(y_pred , y_test)*100) 
```

***now i will explore some hyperparameter which help my accuracy score become optimul***

***hyperparameter of sklearn for RandomForest includes :***

***n_estimators : The n_estimators hyperparameter in a random forest model defines the number of decision trees to build before combining their predictions , by default n_estimators sets on 100***


```python
# let see how n_estimator  our accuracy score of random forest
# STEP:6 MODEL TRAINING AND TESTING
from sklearn.ensemble import RandomForestClassifier
model_1=RandomForestClassifier(n_estimators=200, criterion='entropy',max_depth=10,min_samples_leaf=5,min_samples_split=5)

# TRAIN THE MODEL 
model_1.fit(x_train,y_train)

# ASK THE MODEL TO PREDICT 
y_pred= model_1.predict(x_test)

# EVALUATING THE PERFORMANCE BY COMPARING THE PREDICTED VALUE WITH THE ACTUAL VALUE
from sklearn.metrics import *
print("Accuracy Score of the model is :", accuracy_score(y_pred , y_test)*100) 
```

***The "criterion" in a random forest is the function used to measure the quality of a split at each node of the decision trees within the forest. The algorithm aims to maximize information gain or minimize impurity with each split.***
***it includes Gini Impurity (the default) and Entropy (or Information Gain)***

***max_depth: This parameter sets the maximum number of levels in each decision tree.***

***A shallow tree (low max_depth) may underfit the data, failing to capture complex patterns.***

***A deep tree (high max_depth or default None in scikit-learn, which means unlimited depth until leaves are pure) is more complex and likely to overfit the training data by capturing noise.***

***tuning this helps find the optimal balance between underfitting and overfitting.***

***min_samples_split: This defines the minimum number of samples an internal node must contain before it can be split into a left and right branch.***

**If a node has fewer samples than the specified minimum, the split is not allowed, and the node becomes a leaf node.***

***Increasing this value makes the tree less prone to capturing fine-grained details in the training data, acting as a form of regularization to prevent overfitting. The default value in scikit-learn is 2.***

***min_samples_leaf: This specifies the minimum number of samples that must be present in each resulting leaf node after a split.***

***Even if a potential split satisfies the min_samples_split condition, it will be rejected if it results in any leaf node having fewer samples than min_samples_leaf.***

***This ensures that the final leaf nodes are not based on an arbitrarily small number of samples, providing a stronger guarantee of a minimum number of samples in every leaf than min_samples_split alone. The default value in scikit-learn is 1.***

```python
paramaters_grid = {
    'n_estimators': [100, 200, 500, 1000],
    'max_depth': [10, 20, 30],  # None means trees are grown to full depth
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
model=RandomForestClassifier()
gridsearch=GridSearchCV(estimator=model , param_grid=paramaters_grid)
gridsearch.fit(x_train,y_train)
print(" best parameter values are :", grid_search.best_params_)
```
